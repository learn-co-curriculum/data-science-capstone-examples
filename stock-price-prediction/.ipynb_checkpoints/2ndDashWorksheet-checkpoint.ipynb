```python
def sort_low(x,y):
    if x < y:
        return x
    else:
        return None
    
def sort_high(x,y):
    if x > y:
        return x
    else:
        return None
```

```python
length = 25
scaled_low = [i['Low_1'] for i in predict.stock_dict_list[:length]]
scaled_high = [i['High_1'] for i in predict.stock_dict_list[:length]]
real_low = [i['Real_Low_1'] for i in predict.stock_dict_list[:length] if i['Real_Low_1'] < i['Low_1'] else None]

real_high = [i['Real_High_1'] for i in predict.stock_dict_list[:length] if i['Real_High_1'] < i['High_1'] else None]
x= predict.timestamps[:length]
```
